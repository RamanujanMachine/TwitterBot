```python
from twitter_bot.render_preview import render_preview
from twitter_bot.boinc_result_to_tex import generate_tex_from_str, filename_to_schema
from twitter_bot.get_boinc_results import get_webpage
from typing import Union

def preview(result_filename: str, result: str, schema: Union[str, None] = None) -> None:
    if schema is None:
        schema = filename_to_schema(result_filename)
    with open("result.tex", "w") as result_file:
        result_file.write(generate_tex_from_str(result, schema)[0])
    render_preview("result.tex")

def preview_from_url(url: str, schema: Union[str, None] = None) -> None:
    result_filename = url.split("/")[-1]
    result = get_webpage(url)
    preview(result_filename, result, schema)

```
